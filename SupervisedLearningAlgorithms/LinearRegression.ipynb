```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
data = pd.read_csv("iris.csv")
data.set_index('Id', inplace=True)
data.head()
```

```python
data.info()
```

```python
data.Species = data.Species.apply(lambda s: s[5:])
data
```

```python
data.describe().T
```

```python
feature = data[['PetalLengthCm','PetalWidthCm']]
target = data[['Species']]
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler(feature_range=(-1,1))
scaled = scalar.fit_transform(feature)
feature.iloc[:,0] = scaled[:,0]
feature.iloc[:,1] = scaled[:,1]
feature.head()
```
